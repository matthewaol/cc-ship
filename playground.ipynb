```python
import pandas as pd 
import numpy as np
```

```python
ships_df = pd.read_csv("stem_center_internships.csv")
```

```python
ships_df
```

```python
def find_major(df, major):
    if "computer" in major:
        return df[df['Discipline'].str.contains('Computer|Coding|Multiple',case=False, na=False)]
    if "engineering" in major:
        return df[df['Discipline'].str.contains('Engineering|Multiple',case=False, na=False)]
```

```python
major = str(input("What is your major?"))
find_major(ships_df, major)
```
