```python
import pandas as pd 
import numpy as np
```

```python
ships_df = pd.read_csv("stem_center_internships.csv")
```

```python
ships_df
```

```python
def find_major(major):
    major = major.lower()
    result = ""
    if "computer" in major.lower():
        return ships_df[ships_df['Discipline'].str.contains('Computer|Coding|Multiple', case=False, na=False)]
    elif "engineering" in major.lower():
        return ships_df[ships_df['Discipline'].str.contains('Engineering|Multiple', case=False, na=False)]
    
```

```python
major = str(input("What is your major?"))
results = find_major(major)
results
```

```python
names = results['Internship Opportunity']
links = results['Application Link']
internship_dict = dict(zip(names, links))
internship_dict
```
